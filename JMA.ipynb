## ダウンロードしたCSVデータを読み込む

```python
import os
import sys
import pandas as pd


# 日付，最高気温，最低気温，最深積雪, 降雪量合計, 平均現地気圧, 平均湿度, 平均風速, 最大風速, 風向
# 列の名前を指定する
col_names = ['Date','Temp_Hi','THQ','THX','Temp_Low','TLQ','TLX',
              'Snow_Depth','SDE','SDQ','SDX',
              'Snow_Fall','SFE','SFQ','SFX',
              'Pressure', 'PRQ', 'PRX',
              'Humidity', 'HMQ', 'HMX',
              'Wind_Speed_Av', 'WAQ', 'WAX', 
              'Wind_Speed_Max', 'WMQ', 'Wind_Direction', 'WDQ', 'WDX']

# とりあえず削除する列
removeCols = [u'品質情報', u'均質番号', u'現象なし情報']

path, fname = os.path.split(sys.argv[0])
datadir = os.path.join(os.path.abspath(path), '_csv/')

data = []

# CSVファイルを読み込む
# 上から４行はスキップ
# 'Date'列をパースする
for fname in os.listdir(datadir):
    path = os.path.join(datadir, fname)
    df = pd.read_csv(path, encoding='shift_jis', skiprows=[0,1,2,3], names=col_names, parse_dates=['Date'])

    #　列を選択削除
    for col in df.columns:
        val = df.loc[1, col]
        if val and val in removeCols:
            del df[col]
            
    # 日付をインデックスにする場合
    #df.index = pd.to_datetime(df['Date'])
    #del df['Date']
    data.append(df.dropna())

for df in data:
    print(df.loc[2,'Date'])
```

### 各年度の降雪量の合計，平均，10cm以上の降雪の回数

```python
from datetime import datetime

# 日数を合わせる（12/1〜今日）
delta = datetime.now() - datetime(2014,12,1)
days = delta.days -2

for df in data:
    day = df.loc[2, 'Date']
    snowfall = df.loc[:days, 'Snow_Fall']

    sum_sf = snowfall.sum() # 合計
    ave_sf = snowfall.mean() # 平均
    count_sf = df[df.Snow_Fall > 10] # 10cm以上
    print(day.year, sum_sf, ave_sf, len(count_sf), len(df))
```

## 年度毎の積雪深のグラフを描画

```python
import matplotlib.pyplot as plt
from datetime import datetime

fig, ax = plt.subplots(1,1, figsize=(8,4))
fig.autofmt_xdate(rotation=45, ha='right')

# 日付の範囲
xdates = pd.date_range('20141201', periods=50)

delta = datetime.now() - datetime(2014,12,1)
days = delta.days -2

ax.set_title('Snow Depth')

for df in data:
    ds = df.loc[:days, ['Snow_Depth']]
    
    day = df.loc[2, 'Date']
    ds.columns=[day.year]
    #ds.index = pd.to_datetime(ds['Date'])
    ds.plot(ax=ax)

#fig.savefig('snowdepth.png')
plt.show()
```

### 最新年度の降雪量，積雪深，風速，気温のグラフ

```python
import matplotlib.pyplot as plt

mfig, maxes = plt.subplots(2,1, figsize=(8,8))
mfig.autofmt_xdate(rotation=45, ha='right')

dt = data[-1]

dt['Snow_Depth'].plot(ax=maxes[0], kind='line', color='c')
dt['Snow_Fall'].plot(ax=maxes[0], kind='bar' )

maxes[0].set_title("2014 Snow")
maxes[0].legend(loc='best')

maxes[1].set_title("2014 Temperature")
maxes[1].axhline(y=0.0, linewidth=1, color='k', linestyle='--')

dt['Temp_Hi'].plot(ax=maxes[1], kind='line', color='r')
dt['Temp_Low'].plot(ax=maxes[1], kind='line', color='b')
dt['Wind_Speed_Av'].plot(ax=maxes[1], kind='line', color='g')

plt.grid()
plt.show()
```

### 各年度の降雪量と積雪深のグラフ

```python
from matplotlib import dates

def plotData(df, axn):
    axn.set_ylabel('Fall [cm]')
    axn.set_ylim(0, 50)
    axn2 = axn.twinx()
    axn2.set_ylabel('Depth [cm]')
    axn2.set_ylim(0, 150)
    
    df['Snow_Fall'].plot(ax=axn, kind='bar' )
    df['Snow_Depth'].plot(ax=axn2, kind='line', color='c')

n = len(data)
fig, axes = plt.subplots(n,1, figsize=(8,8))
plt.subplots_adjust(hspace=0.3)
fig.autofmt_xdate(rotation=45, ha='right')

days = dates.WeekdayLocator()
daysFmt = dates.DateFormatter('%m/%d')
axes[n-1].xaxis.set_major_locator(days)
axes[n-1].xaxis.set_major_formatter(daysFmt)

for i in range(n):
    df = data[i]
    axes[i].set_title( df.loc[2,'Date'].year)
    axes[i].axhline(y=10)
    plotData(df, axes[i])

plt.show()
```
